```python
import numpy as np
import matplotlib.pyplot as plt

# Test importu topfarm
import topfarm
print("TopFarm zaimportowany – wersja:", topfarm.__version__)

# Dane turbin – przykład z zadania
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

# Rysowanie wykresu
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=180, color='darkblue', marker='o', edgecolor='black')
plt.title("Layout turbin – Zadanie 2.1 (pierwszy wykres)", fontsize=14)
plt.xlabel("x [m]", fontsize=12)
plt.ylabel("y [m]", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')
plt.tight_layout()
plt.show()
import topfarm
print("TopFarm zaimportowany – wersja:", topfarm.__version__)
```

```python
# Ulepszony wykres layoutu turbin
# Zakładamy, że zmienne `x` i `y` są już zdefiniowane w poprzedniej komórce
colors = ['blue', 'red', 'green', 'orange']
labels = ['T1', 'T2', 'T3', 'T4']
plt.figure(figsize=(8, 6))
for xi, yi, c, lab in zip(x, y, colors, labels):
    plt.scatter(xi, yi, s=220, color=c, marker='o', edgecolor='black', linewidth=0.8)
    # Etykieta nieco obok i powyżej punktu
    plt.text(xi + 30, yi + 25, lab, fontsize=12, fontweight='bold', va='bottom', ha='left')
plt.title("Layout turbin – Zadanie 2.1 (ulepszony wykres)", fontsize=18)
plt.xlabel("x [m]", fontsize=14)
plt.ylabel("y [m]", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')
plt.legend(title='Turbiny')
plt.tight_layout()
plt.show()
```
